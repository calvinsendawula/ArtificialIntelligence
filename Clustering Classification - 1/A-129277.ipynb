```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

warnings.filterwarnings('ignore')
df = pd.read_csv('Clustering_Data.csv')
df.head(4)
```

```python
df
```

```python
scaler = StandardScaler()   
scaled_df = scaler.fit_transform(df)

kmeans = KMeans(n_clusters=4)

kmeans.fit(scaled_df)

print(kmeans.cluster_centers_)
```

```python
centroids = kmeans.cluster_centers_
```

```python
centroids[3]
```

```python
print(kmeans.labels_)
```

```python
df['labels'] = kmeans.labels_
df.head(4)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df.head(4)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.3)

train_x = train.drop('labels', axis=1)
train_y = train['labels']

test_x = test.drop('labels', axis=1)
test_y = test['labels']
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=7, metric='euclidean')

knn_clf.fit(train_x, train_y)
```

```python
pred = knn_clf.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy of the classifier:", accuracy_score(test_y, pred))
```
