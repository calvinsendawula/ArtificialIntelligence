```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

warnings.filterwarnings('ignore')

# Question a solution begins here
# Reading the data to be used to generate the clusters
df = pd.read_csv('Clustering_Data.csv')
df.head()
```

```python
df
```

```python
# performing scaling
scaler = StandardScaler()   
scaled_df = scaler.fit_transform(df)

# Creating 4 clusters (k=4)
kmeans = KMeans(n_clusters=4)

# Fit the model
kmeans.fit(scaled_df)

# Get the cluster centroids
print(kmeans.cluster_centers_)
```

```python
centroids = kmeans.cluster_centers_
```

```python
centroids[3]
```

```python
print(kmeans.labels_)
```

```python
# Add the labels to the dataframe
df['labels'] = kmeans.labels_
df.head()
```

```python
df2 = pd.read_csv('Classification_Data.csv')
df2.head(5)
```

```python
# Shuffle the dataset
df = df.sample(frac=1).reset_index(drop=True)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

acc = 0

# train the model until the accuracy is greater than 0.9
while acc < 0.9:
    # generate a test size randomly to 2 decimal places
    rand = np.random.uniform(0.2, 0.3)
    rand = round(rand, 2)
    
    # Split the dataset into training and testing sets
    train, test = train_test_split(df, test_size=rand)
    
    # Create features and labels(targets)
    train_x = train.drop('labels', axis=1)
    train_y = train['labels']

    test_x = test.drop('labels', axis=1)
    test_y = test['labels']

    # create KNN classifier object
    knn_clf = KNeighborsClassifier(n_neighbors=7, metric='euclidean')

    # train the model(object) using the training sets
    knn_clf.fit(train_x, train_y)

    # predict the response for test dataset
    pred = knn_clf.predict(test_x)
    
    # Model Accuracy, how often is the classifier correct?
    acc = accuracy_score(test_y, pred)
print("Accuracy of the classifier:", acc)
print("Test size:", rand)
```
