```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Filter out warnings
warnings.filterwarnings('ignore')

# Read the data where the clusters are going to come from
df = pd.read_csv('Clustering_Data.csv')
df.head()
```

```python
# Print out the dataframe (df.shape prints out the number of rows and columns)
# df.shape
df
```

```python
# perform scaling on the data so the cluster values are not skewed too far apart
scaler = StandardScaler()   
scaled_df = scaler.fit_transform(df)

# Create 4 clusters
kmeans = KMeans(n_clusters=4)

# Fit the model using the scaled data
kmeans.fit(scaled_df)

# Get the cluster centroids
print(kmeans.cluster_centers_)
```

```python
# keep the centroids in a variable
centroids = kmeans.cluster_centers_
```

```python
# print the last row of the centroids (don't have to do this)
centroids[3]
```

```python
# Get the labels
print(kmeans.labels_)

# we have 4 unique labels (0, 1, 2, 3), matching the 4 clusters we created
```

```python
# Add the labels to the dataframe
df['labels'] = kmeans.labels_
df.head()
```

```python
# Here we are reading the classification data which we will predict labels for later
df2 = pd.read_csv('Classification_Data.csv')
df2.head(5)
```

```python
# Shuffle the dataset to make the data more random
df = df.sample(frac=1).reset_index(drop=True)
df.head(5)
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.3)

# Create features and labels(targets)
train_x = train.drop('labels', axis=1)
train_y = train['labels']

test_x = test.drop('labels', axis=1)
test_y = test['labels']
```

```python
# create KNN classifier object
knn_clf = KNeighborsClassifier(n_neighbors=7, metric='euclidean')

# train the model(object) using the training sets
knn_clf.fit(train_x, train_y)
```

```python
# predict the response for test dataset
pred = knn_clf.predict(test_x)
```

```python
# Model Accuracy, how often is the classifier correct?
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(test_y, pred))
```
